```python
from matplotlib import style
```

```python
x = [4,5,1,8,5]
y = [4,5,6,1,4]
plt.plot(x,y)
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [4,5,1,8,5]
y = [4,5,6,1,4]
style.use("ggplot")
plt.plot(x,y)
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [4,5,1,8,5]
y = [4,5,6,1,4]
x1 = [1,2,5,4,7]
y1 = [4,5,6,8,7]
style.use("ggplot")
plt.plot(x,y,'g',label = 'LineOne',linewidth=5)
plt.plot(x1,y1,'c',label = 'Linetwo',linewidth=5)
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [4,5,1,8,5]
y = [4,5,6,1,4]
x1 = [1,2,5,4,7]
y1 = [4,5,6,8,7]
style.use("ggplot")
plt.plot(x,y,'g',label = 'LineOne',linewidth=5)
plt.plot(x1,y1,'c',label = 'Linetwo',linewidth=5)
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show
```
