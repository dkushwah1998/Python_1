# Box plot

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\HP\Desktop\Demos Files\Day 10 - Demos\cancer.csv')
df
```

```python
df.columns
```

```python
plt.boxplot(df.radius_mean)
```

```python
plt.boxplot(df.radius_mean, showmeans =True, vert=False, sym = 'b+')
plt.show()
```

```python
## using pandas
df.boxplot(column = 'area_mean', by = 'diagnosis');
plt.title('')
plt.show()
```

```python
sns.boxplot(x='diagnosis', y = 'area_mean',data = df)
```

```python
malignant = df[df['diagnosis']=='M']['area_mean']
benign = df[df['diagnosis']=='B']['area_mean']

fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot([malignant,benign], labels=['M', 'B'])
```

# Scatterplot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
tips = pd.read_csv('C:/Users/HP/Desktop/Demos Files/Day 10 - Demos/tips.csv')
tips
```

```python
tips.describe()
```

```python
sns.lmplot(x='total_bill',y = 'tip', data = tips)
```

```python
sns.lmplot(x='total_bill',y = 'tip',hue = 'sex', data = tips)
```

```python
sns.lmplot(x='total_bill',y = 'tip',hue = 'sex', data = tips, markers = ['o','x'])
```

```python
sns.lmplot(x='total_bill',y = 'tip',col = 'sex', data = tips)
```

```python
sns.lmplot(x='total_bill',y = 'tip',row = 'smoker',col = 'sex', data = tips)
```

# Heatmaps

```python
car = pd.read_csv(r'C:\Users\HP\Desktop\Demos Files\Day 10 - Demos\cars.csv')
```

```python
car
```

```python
f,ax = plt.subplots(figsize = (20,15))
sns.heatmap(car.corr(),center = 0,cmap = 'Reds')
ax.set_title('Multi collinearity of car')
```

```python
f,ax = plt.subplots(figsize = (20,15))
sns.heatmap(car.corr(),center = 0,cmap = 'Reds', annot = True)
ax.set_title('Multi collinearity of car')
```
