# Pie chart

```python
import matplotlib.pyplot as plt
```

```python
language = 'python','java','c++','PHP','JavaScript'
popularity = [20,23.56,23.5,45.8,6.84]
```

```python
plt.pie(popularity, labels = language)
plt.title('Poplarity of languege')
plt.show()
```

```python
plt.pie(popularity, labels = language,autopct = '%1.2f%%')
plt.title('Poplarity of languege')
plt.show()
```

```python
explode = (0,0,0.2,0,0)
plt.pie(popularity, labels = language,autopct = '%1.2f%%', explode = explode)
plt.title('Poplarity of languege')
plt.show()
```

```python
plt.pie(popularity, labels = language,autopct = '%1.2f%%', explode = explode,shadow = True)
plt.title('Poplarity of languege')
plt.show()
```

```python
## start angle rotate the pie chart
plt.pie(popularity, labels=language, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140)
plt.show()
```

```python
plt.pie(popularity, labels=language, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140, frame = True)
plt.show()
```

```python
plt.pie(popularity, labels=language, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140, frame = True,
        rotatelabels=True)
plt.show()
```

```python
plt.pie(popularity, labels=language, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140, frame = True,
        rotatelabels=True, wedgeprops = {'linewidth':2})
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(popularity, labels=language, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140, frame = True,
        rotatelabels=True, wedgeprops = {'linewidth':2}, radius = 3)
plt.show()
```

# Histogram

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
values = [10,15,13,20,14,15,24,26,34,42,36,29,24,15,20,16,11,8]
plt.hist(values, histtype = 'bar')
```

```python
plt.hist(values,bins = 10,histtype = 'bar', edgecolor = 'k')
plt.show()
```

```python
plt.hist(values,bins = 10, edgecolor = 'k')
plt.show()
```

```python
plt.hist(values,bins = 10, edgecolor = 'k')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(values,bins = 10, edgecolor = 'k',label = 'Test Score Data')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.hist(values,bins = 10, edgecolor = 'k',label = 'Test Score Data',range = (0,60))
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.hist(values,bins = 10, edgecolor = 'k',label = 'Test Score Data',orientation = 'horizontal')
plt.title('Histogram')
plt.ylabel('Values')
plt.xlabel('Frequency')
plt.legend()
plt.show()
```

```python
values1 = [50,15,25,24,56,4,65,8,78,4,8,56,85,4,5,78]

plt.hist(values,bins = 10, edgecolor = 'k',label = 'Test1 Score Data')
plt.hist(values1,bins = 10, edgecolor = 'k',label = 'Test2 Score Data')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.hist(values,bins = 10, edgecolor = 'k',label = 'Test1 Score Data',alpha = 0.4)
plt.hist(values1,bins = 10, edgecolor = 'k',label = 'Test2 Score Data',alpha = 0.5)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.hist(values,bins = 10, edgecolor = 'k',label = 'Test1 Score Data',alpha = 0.4,color ='c')
plt.hist(values1,bins = 10, edgecolor = 'k',label = 'Test2 Score Data',alpha = 0.5,color = 'r')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
