# Problem Statement 2: Write a Python program to perform the following tasks-1.Create a pandasseries from the below dictionary where indices are subjects:

```python
data = {'English':{'Sam':60,'Jackson':74,'Ahree':85}, 'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
      'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
      'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
      'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}}
series = pd.Series(data)
series

```

# 2.Convert the created series into DataFrame and replace the null values with zeroes.

```python
import numpy as np
import pandas as pd
data1 =[{'Sam':60,'Jackson':74,'Ahree':85},{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
      {'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
      {'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
      {'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}]
df = pd.DataFrame(data1, index = ['English','History','Geography','Mathematics','Science'])

df
```

```python
df.isnull()
```

```python
df = df.fillna(0)
df = df.astype(int)
df
```

# 3.Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects.

```python
## Transpose 
df_transpose = df.transpose()  ## OR df.T()
df_transpose
```

```python
df_transpose["Average Marks"] = (df_transpose['English'] +df_transpose['History'] + df_transpose['Geography'] + 
                          df_transpose['Mathematics'] + df_transpose['Science'])/5
df_transpose
```

# Problem Statement 3: Write a Python program to create a series from 1 to 1000 and select only numbers divisible by 7 and 17

```python
import numpy as np
import pandas as pd
lst = []

for i in range(1,1000):
    if i%7==0 and i%17==0:
        lst.append(str(i))
print(','.join(lst))
```

```python
series = pd.Series(lst)
series
```
