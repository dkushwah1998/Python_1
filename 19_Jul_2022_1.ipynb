```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
## Standard Scaler
data1 = np.array([[-100.3],[27.5],[0],[-200.9],[1000]])
print("Before Scaling:\n",data1)
standard_scale = preprocessing.StandardScaler()
scaled = standard_scale.fit_transform(data1)
print("\nAfter Standard Scaler:\n",scaled)
```

```python
#MinMax Scaler
data1 = np.array([[-100.3],[27.5],[0],[-200.9],[1000]])
print("Before Scaling:\n",data1)
min_max_scaler = preprocessing.MinMaxScaler(feature_range = (1,2))
scaled = min_max_scaler.fit_transform(data1)
print("\nAfter Scaler:\n",scaled)
```

```python
## RobustScaler()
data1 = np.array([[-100.3],[27.5],[0],[-200.9],[1000]])
print("Before Scaling:\n",data1)
robust_scaler = preprocessing.RobustScaler()
scaled = robust_scaler.fit_transform(data1)
print("\nAfter Scaler:\n",scaled)
```

```python
data = np.array([[5,4.1,3.2,1.2,.8],[5.3,4.6,3.5,1.5,0.6],
                [5.6,4.2,3.4,1.2,0.4],[5.8,4.6,3.1,1.7,0.4]])
print("Before nomalisation:\n",data)
normalized_data = preprocessing.normalize(data)
print("\nAfter Normalization\n",normalized_data)
```
