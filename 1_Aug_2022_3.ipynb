# 1. Margin of error and confidence intervel

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.utils import shuffle
```

```python
boston = load_boston()
x, y = shuffle(boston.data, boston.target,random_state = 13)
```

```python
df = pd.DataFrame(boston.data,columns = boston['feature_names'])
df['target'] = boston.target
df
```

```python
## Generate sample
sample_size = 200
sample = df.sample(n = sample_size, random_state = 1)
sample
```

```python
import math
from scipy import stats
```

```python
sample_mean = sample.target.mean()
np.random.seed(1)
z_critical = stats.norm.ppf(q = 0.975)  
pop_stdev = sample.target.std()
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error) 
print(z_critical)
print(margin_of_error)
print(confidence_interval)
```

# 2. Hypothesis Test

```python
import numpy as np 
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.utils import shuffle
```

```python
boston = load_boston()
x, y = shuffle(boston.data, boston.target, random_state = 13)
```

```python
df = pd.DataFrame(boston.data,columns = boston['feature_names'])
df.head()
```

```python
df['target'] = boston.target
pop_target = df.target
mu = pop_target.mean()
sigma = pop_target.std()
print('Population mean: ',mu,'\nPopulation Satandard deviation: ',sigma)
```

```python
sample_size = 100
sample = df.sample(n = sample_size, random_state = 4).target
sample_mu = sample.mean()
sample_sigma = sample.std()
print('sample mean: ',sample_mu, '\nsSampel Standard deviation: ',sample_sigma)
```

```python
## null hyposthesis: population mean = sample mean
import math
from scipy import stats
z_critical = stats.norm.ppf(q = 0.975)
SE = sigma/np.sqrt(sample_size)
z_stat = (sample_mu-mu)/SE
print('Z-critical: ',z_critical, "\nZ-Statistics: ",z_stat)
```

```python
## One Sample test
stats.ttest_1samp(a =sample,popmean = pop_target.mean())
```

```python
sigma = sample.std()/math.sqrt(100)
stats.t.interval(0.95,loc = sample.mean(), df=45,scale = sigma)
```

```python
## Two sample test
sample2 = df.sample(n=100,random_state =88).target
sample2_mu = sample2.mean()
sample2_std = sample2.std()
print("sample mean: ",sample2_mu,'\nsample2 std: ', sample2_std)
```

```python
print(stats.ttest_ind(a= sample,
                b= sample2,
                equal_var=False))  
```
