```python
df3 = pd.read_csv('C:/Users/HP/Desktop/Python 1/Day 6 - Demos/diabetes_data2.csv', index_col = 'PatientID')
print(df3.shape)
df3.head(5)
```

```python
data2 = pd.concat([data1,df3], axis = 1)
print(data2.shape)
data2
```

```python
df_left = pd.merge(data1, df3, on = 'PatientID',how = 'left')
df_left
```

```python
df_outer = pd.merge(data1, df3, on = 'PatientID',how = 'outer')
df_outer
```

```python
df_inner = pd.merge(data1, df3, on = 'PatientID',how = 'inner')
df_inner
```
